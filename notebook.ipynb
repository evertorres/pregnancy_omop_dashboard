```python
import os
import sqlalchemy
from sqlalchemy import create_engine
import pandas as pd

# Forzar a que el cliente de Postgres use UTF8
os.environ['PGCLIENTENCODING'] = 'utf-8'

# Si el error persiste, a veces es por una variable de sistema llamada 'LC_ALL' o 'LANG'
os.environ['PYTHONIOENCODING'] = 'utf-8'
```

```python
db_url = 'postgresql://admin:1320@localhost:5432/10k_pregnant_db'
engine = create_engine(
    db_url,
    connect_args={
        'client_encoding': 'utf8',
        'options': '-c client_encoding=utf8'
    }
)
```

```python
query = """
        SELECT person.gender_concept_id, concept.concept_name, count(person.gender_concept_id) as total
        FROM cdm_synthea10.person
        JOIN cdm_synthea10.concept on concept.concept_id = person.gender_concept_id
        GROUP BY person.gender_concept_id, concept.concept_name;
        """
    
df = pd.read_sql(query, engine)
```
